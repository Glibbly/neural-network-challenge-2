## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
column_list = ['Age', 'MaritalStatus', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'YearsAtCompany']

# Create X_df using your selected columns
X_df = attrition_df[column_list]
X_df.head()

# Show the data types for X_df
X_df.dtypes

#Im performing get_dummies here to get numeric values in my columns
X_df = pd.get_dummies(X_df, dtype = int)
X_df.head()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
#done above
X_train.head()
```

```python
# Create a StandardScaler
ss = StandardScaler()

# Fit the StandardScaler to the training data
ss_fit = ss.fit(X_train)

# Scale the training and testing data
ss_transformed_train = ss_fit.transform(X_train)
ss_transformed_test = ss_fit.transform(X_test)

ss_transformed_train
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop = None)

# Fit the encoder to the training data
ohe.fit(np.array(y_train["Department"]).reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_department_ohe_train = ohe.transform(np.array(y_train["Department"]).reshape(-1, 1))
y_department_ohe_test = ohe.transform(np.array(y_test["Department"]).reshape(-1, 1))

y_department_ohe_train
y_department_ohe_test
```

```python
# Create a OneHotEncoder for the Attrition column
ohe2 = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop = None)

# Fit the encoder to the training data
ohe2.fit(np.array(y_train["Attrition"]).reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_attrition_ohe2_train = ohe2.transform(np.array(y_train["Attrition"]).reshape(-1, 1))
y_attrition_ohe2_test = ohe2.transform(np.array(y_test["Attrition"]).reshape(-1, 1))
y_attrition_ohe2_train
y_attrition_ohe2_test
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data.
ss_transformed_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(ss_transformed_train.shape[1],), name='input_layer')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu', name="shared_layer1")(input_layer)
shared_layer2 = layers.Dense(32, activation='relu', name="shared_layer2")(shared_layer1)
ss_transformed_train.shape[1]
```

```python
# Create a branch for Department
# with a hidden layer and an output layer


# Create the hidden layer
dept_hidden_layer1 = layers.Dense(32, activation='relu', name="dept_hidden_layer1")(shared_layer2)

# Create the output layer
department_output = layers.Dense(y_department_ohe_train.shape[1], activation='softmax', name="department_output")(dept_hidden_layer1)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden_layer2 = layers.Dense(32, activation='relu', name="attrition_hidden_layer2")(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(y_attrition_ohe2_train.shape[1], activation='softmax', name="attrition_output")(attrition_hidden_layer2)
```

```python
# Create the model
nn = Model(inputs=input_layer, outputs={"department_output" : department_output, "attrition_output" : attrition_output})

# Compile the model
nn.compile(optimizer='adam',
              loss={
                  'department_output': 'categorical_crossentropy',
                  'attrition_output': 'binary_crossentropy'
              },
              metrics={
                  'department_output': 'accuracy',
                  'attrition_output': 'accuracy'
              })

# Summarize the model
nn.summary()
```

```python
print(y_attrition_ohe2_train.dtype)
```

```python
# Train the model
fit_model = nn.fit(ss_transformed_train, {"department_output" : y_department_ohe_train, "attrition_output" : y_attrition_ohe2_train}, epochs=50, batch_size=32)
```

```python
# Evaluate the model with the testing data
fit_model.predict(ss_transformed_test, {"department_output" : y_department_ohe_test, "attrition_output" : y_attrition_ohe2_test})
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {fit_model.history['department_output_accuracy'][-1]}")
print(f"Attrition Accuracy: {fit_model.history['attrition_output_accuracy'][-1]}"
```
